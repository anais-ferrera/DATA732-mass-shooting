```python
import pandas as pd
```

```python
df=pd.read_csv('data_us.csv')
df
```

```python
df=df.drop(df.columns[[10,12,14,18,19,22,23]], axis='columns')
```

```python
df=df.rename(columns = {'location.1': 'place'})
```

```python
df
```

```python
df[['city','state']] = df.location.str.split(",",expand=True,)
df['city']=df.city.str.strip()
df['state']=df.state.str.strip()
```

```python
df
```

```python
df.to_csv('dataFinalUS.csv')
```

```python
df
```

```python
dfAge=df[df['age_of_shooter']!='-']
dfAge
```

```python
dfAge["age_of_shooter"] = pd.to_numeric(dfAge["age_of_shooter"])
```

```python
listeValue=[]
```

```python
num15=len(dfAge.loc[(dfAge['age_of_shooter']<=15),:])
listeValue.append(num15)
```

```python
num1621=len(dfAge.loc[(dfAge['age_of_shooter']>=16) & (dfAge['age_of_shooter']<=21),:])
listeValue.append(num1621)
```

```python
num2230=len(dfAge.loc[(dfAge['age_of_shooter']>=22) & (dfAge['age_of_shooter']<=30),:])
listeValue.append(num2230)
```

```python
num3140=len(dfAge.loc[(dfAge['age_of_shooter']>=31) & (dfAge['age_of_shooter']<=40),:])
listeValue.append(num3140)
```

```python
num4150=len(dfAge.loc[(dfAge['age_of_shooter']>=41) & (dfAge['age_of_shooter']<=50),:])
listeValue.append(num4150)
```

```python
num51=len(dfAge.loc[(dfAge['age_of_shooter']>=51),:])
listeValue.append(num51)
```

```python
listeValue
```

```python
df2=pd.DataFrame(columns=['categorie','value'])
df2
```

```python
df2['categorie']=['<15','16-21','22-30','31-40','41-50','>51']
df2['value']=listeValue
df2
```

```python
df2.to_csv('dataAge.csv')
```
